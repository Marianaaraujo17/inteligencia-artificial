### Importando as bibliotecas

```python
# Importa√ß√£o de Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')
```

### Lendo o arquivo

```python
# leitura de dados (CSV), separados por ponto e v√≠rgula no nosso caso
df = pd.read_csv('house.csv',';')
df
```

### Separando dados de Treinamento (70%) e Testes (30%)

```python
# separa a coluna target (colunas com os dados das classes)
y = 'sellingPrice'

# separa as colunas de features (colunas com os dados)
x = df.drop(columns=['sellingPrice'])

y = df[y]

print("X: \n", x)
print("Y: \n", y)
```

```python
# Uso do Modelo para fazer Classifica√ß√µes com os dados de Testes
# Para funcionar, deve ser importada a biblioteca:
from sklearn.model_selection import train_test_split
```

```python
# Separa√ß√£o de dados para Treinamento (train) e Testes (test); Vari√°veis X s√£o as colunas features e Vari√°veis Y trata-se da coluna target
Xtrain, Xval, Ytrain, Yval = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
# Visualiza√ß√£o dos dados de Treinamento e Testes, j√° separados
Xtrain
Xval
```

### Regress√£o Linear

```python
# Cria√ß√£o do modelo de Regress√£o linear
model = LinearRegression()
```

```python
# Treino do modelo com os Dados
model.fit(Xtrain, Ytrain)
```

#### Coeficiente de Determina√ß√£o R2

```python
# C√°lculo do Coeficiente de Determina√ß√£o R2 (quanto mais pr√≥ximo de 1, melhor o modelo!)
r_sq = model.score(Xtrain, Ytrain) 
print('Coefficient of determination:', r_sq)
```

#### C√°lculo de MSE (Mean Squared Error)

```python
# C√°lculo de MSE (Mean Squared Error)
np.mean((model.predict(Xval)-Yval)**2)
```

#### Regress√£o Linear: ùëì(ùë•) = ùëè‚ÇÄ + ùëè‚ÇÅùë•

```python
# Admitindo a Equa√ß√£o da Regress√£o Linear: ùëì(ùë•) = ùëè‚ÇÄ + ùëè‚ÇÅùë•
print('b0 ou intercept:', model.intercept_)
print('b1 ou coeficiente (slope):', model.coef_)
```
